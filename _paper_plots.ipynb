```python
import os
import numpy             as np
import matplotlib.pyplot as plt
import pandas            as pd 
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)


folder = './plots/paper/'
if not os.path.exists(folder):
    os.makedirs(folder)
```

```python
our_color = '#F2BB66'
pjv_color= '#779ECB'
```

```python
our_marker = 'D'
pjv_marker = 's'
```

```python
marker_range_10 = range(0,10,1)
marker_range_100 = range(0,100,10)
```

```python
# JWT Length of VPs when iterating over permissions. Each credential was delegated 10 times. 
# This graph plots the length of the 10th delegated credential when the number of permission varies between 1 and 10.
# All permissions included in the VC are disclosed.
# Delegators = 10, Permissions = [1...10]
df = pd.read_csv("./csv_dir/iop_vp_jwt_length.csv")
fig, ax = plt.subplots()
x = range(1,11,1)

ax.plot(x, df['Ours'] / 1000, label='Ours', color=our_color, marker=our_marker, markevery=marker_range_10, linewidth=2)
ax.plot(x, df['PJVs'] / 1000, label='Vrielynck et al.', color=pjv_color, marker=pjv_marker, markevery=marker_range_10, linewidth=2)

ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.yaxis.set_major_locator(MultipleLocator(5))
ax.yaxis.set_minor_locator(AutoMinorLocator(5))

ax.set_xlabel("Number of permissions")
ax.set_ylabel("VP Size (KB)")

ax.legend(loc="upper left", fancybox=True, framealpha=0.4, prop={'size': 9})

plt.xlim(0.9, 10.1)
plt.ylim(0, 50)

plt.grid(True, which='both', linestyle=':', linewidth=0.5)
plt.savefig(f"{folder}/Iterating_Over_Permissions_JWT_Length.pdf", dpi=600, format='pdf')
plt.show()
```

```python
# JWT Length of VPs. Each VP is formed from a VC that was obtained by delegation. The first credential has 10
# permissions and delegates 9 of them. The second credential has 9 permissions and delegates 8 of them...
# All permissions disclosed.
# Delegators = 10, Permissions = [10...1]
df = pd.read_csv("./csv_dir/rp_vp_jwt_length.csv")
fig, ax = plt.subplots()
x = range(1,11,1)

ax.plot(x, df['Ours'] / 1000, label='Ours', color=our_color, marker=our_marker, markevery=marker_range_10, linewidth=2)
ax.plot(x, df['PJVs'] / 1000, label='Vrielynck et al.', color=pjv_color, marker=pjv_marker, markevery=marker_range_10, linewidth=2)

ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.yaxis.set_major_locator(MultipleLocator(5))
ax.yaxis.set_minor_locator(AutoMinorLocator(5))

ax.set_xlabel("Delegation chain length")
ax.set_ylabel("VP Size (KB)")

ax.legend(loc="upper left", fancybox=True, framealpha=0.4, prop={'size': 9})

plt.xlim(0.9, 10.1)
plt.ylim(0, 45)

plt.grid(True, which='both', linestyle=':', linewidth=0.5)
plt.savefig(f"{folder}/Retain_Permissions_JWT_Length.pdf", dpi=600, format='pdf')
plt.show()
```

```python
# JWT Length of VPs. Each VP is formed from a VC that was obtained by delegation. Each VC that is obtained by 
# delegation has 10 permissions and is delegated from delegator to delegatee 10 times as follows:
#      Delegator D0 -> Delegatee D1
#      Delegator D1 -> Delegatee D2
#      ...
#      Delegator D9 -> Delegatee D10
# The graph plots the VPs generated from each of these credentials in which the presenter discloses 10/10 permissions
# Delegators = [1...10], Permissions = 10, Disclosures = 10
df = pd.read_csv("./csv_dir/10_iod_vp_jwt_length.csv")
fig, ax = plt.subplots()
x = range(1,11,1)

ax.plot(x, df['Ours'] / 1000, label='Ours', color=our_color, marker=our_marker, markevery=marker_range_10, linewidth=2)
ax.plot(x, df['PJVs'] / 1000, label='Vrielynck et al.', color=pjv_color, marker=pjv_marker, markevery=marker_range_10, linewidth=2)

ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.yaxis.set_major_locator(MultipleLocator(5))
ax.yaxis.set_minor_locator(AutoMinorLocator(5))

ax.set_xlabel("Delegation chain length")
ax.set_ylabel("VP Size (KB) [10/10 Permissions Disclosed]")

ax.legend(loc="upper left", fancybox=True, framealpha=0.4, prop={'size': 9})

plt.xlim(0.9, 10.1)
plt.ylim(0, 45)

plt.grid(True, which='both', linestyle=':', linewidth=0.5)
plt.savefig(f"{folder}/Iterating_Over_Delegators_10_Disclosures_JWT_Length.pdf", dpi=600, format='pdf')
plt.show()
```
